```python
# Basic imports
import torch
from torch import nn

from assets.shapes import square, circle, stanford_bunny
from assets.loss_functions import mse
from assets.utils import automate_training, plot_curves
```

# Showcase

## Heart

```python
from assets.shapes import heart
from assets.networks import NIG_Net

# Generate target curve points
num_pts = 1000
t = torch.linspace(0, 1, num_pts).reshape(-1, 1)
X_t = heart(num_pts)

nig_net = NIG_Net(
    layer_count_inj = 4, smm_num_grps = 6, smm_neurons_per_grp = 6
)
automate_training(
    model = nig_net, loss_fn = mse, X_train = t, Y_train = X_t,
    learning_rate = 0.001, epochs = 10000, print_cost_every = 2000
)

X_p = nig_net(t)
plot_curves(X_p, X_t)
```

## Puzzle Piece

```python
from assets.shapes import puzzle_piece
from assets.networks import NIG_Net

# Generate target curve points
num_pts = 1000
t = torch.linspace(0, 1, num_pts).reshape(-1, 1)
X_t = puzzle_piece(num_pts)

nig_net = NIG_Net(
    layer_count_inj = 5, smm_num_grps = 15, smm_neurons_per_grp = 15
)
automate_training(
    model = nig_net, loss_fn = mse, X_train = t, Y_train = X_t,
    learning_rate = 0.1, epochs = 1000, print_cost_every = 200
)

X_p = nig_net(t)
plot_curves(X_p, X_t)
```

Now we try to fit the puzzle piece using a Pre-Auxilliary network augmented NIG_Net.

```python
from assets.shapes import puzzle_piece
from assets.networks import PreAux_NIG_Net

# Generate target curve points
num_pts = 1000
t = torch.linspace(0, 1, num_pts).reshape(-1, 1)
X_t = puzzle_piece(num_pts)

preaux_nig_net = PreAux_NIG_Net(
    layer_count_inj = 5, smm_num_grps = 15, smm_neurons_per_grp = 15,
    layer_count_preaux = 2, hidden_dim_preaux = 25
)
automate_training(
    model = preaux_nig_net, loss_fn = mse, X_train = t, Y_train = X_t,
    learning_rate = 0.1, epochs = 10000, print_cost_every = 2000
)

X_p = preaux_nig_net(t)
plot_curves(X_p, X_t)
```
